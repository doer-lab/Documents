# 一、文件操作

### 使用系统自带open()函数，还有一个模块 codecs 可实现文件的读取，且更方便
### 均以实际例子进行讲解

## 1、从文件中读取数据——将文件读取到内存后，就可以对其进行操作了

### 1.1 读取整个文件

#### 要读取一个文件，需要一个包含几行文本的文件。在此，我们以当前目录下存在的 test_data.txt 文件（其内部包含100条数据）作为操作对象，进行读取其内容的操作

```python
with open('test_data.txt',encoding='utf-8') as file:
    contents = file.read()
    print(contents)
```

### 【说明】：
#### 1、要以任何方式使用文件，都必须先将该文件打开，这样做才能访问它
#### 2、open()接受的第一个参数：要打开的文件的名称，用以读取当前工作路径(就是该程序所在的文件目录)下的 test_data.txt 文件，若当先工作路径下不存在指定的文件，则会报错；第二个参数是 encoding 表示要读取的文件的编码格式，若编码格式错误，同样会报错。【注：第一个参数可带路径】
#### 3、函数open()返回的是一个表示文件的对象。在上述的代码中，其返回的是一个表示 test_data.txt 的文件对象，Python将这个对象存储在file中，以后我们后面使用它
#### 4、关键字 with 在不再需要访问文件后将自动将文件关闭。可通过open()和close()来读取文件，这样做会比较麻烦，且会引发一些其他的问题，推荐使用 with 来进行文件的读取
#### 5、有了表示 test_data.txt 文件的对象后，我们用read()方法读取这个文件的全部内容，并将其作为一个长长的字符串存储在变量 contents 中
#### 6、最后通过print()函数将存储在变量 contents 中的所有内容打印出来，也即显示 test_data.txt 文件中的内容

### 1.2 文件路径

#### 要让Python打开不与程序文件处于同一个目录下的文件，需要提供文件路径（可以是相对路径，也可以是绝对路径），让Python到系统指定的特定位置去查找指定的文件。ex: 读取test_data_1.txt文件，经查，它在当前工作路径中的python_work的文件夹下

#### ①使用相对路径

```python
with open('python_work/test_data_1.txt',encoding='utf-8') as file_1:
    contents_1 = file_1.read()
    print(contents_1)
```

#### ②使用绝对路径

```python
with open('E:/Competition/TensorflowLearning/文档整理/python_work/test_data_1.txt',encoding='utf-8') as file_3:
    contents_3 = file_3.read()
    print(contents_3)
```

#### 从②可看出，带绝对路径的文件名称比较长，我们可先将该路径存储在一个单独的变量中，在向open()函数中传入表示指定文件名称的变量；使用绝对路径，我们可读取任何地方的文件，一般不建议使用绝对路径

### 1.3 逐行读取数据

#### 要逐行读取数据，可对文件对象使用 readlines() 方法。readlines() 从文件中读取每一行，并将读取的所有数据存储在一个列表中。以实际例子进行演示：

#### 此外，还有readline()函数实现对文本数据的读取，该函数每次仅读取一行数据便返回，即：返回的数据只是原始数据中的某一行

```python
with open('test_data.txt',encoding='utf-8') as file_obj:    # file_obj 只可在该代码块内部进行使用
    content_lists = file_obj.readlines()        # content_lists 可在该代码块外部进行使用
    print(content_lists)
```

```python
print('The type of content_lists is ',type(content_lists))              # 查看content_list 的类型，得知其是一个 list 的数据类型
```

#### 【注】: 读取文本文件时，Python将其中的所有文本都解读为字符串。如果我们读取的数字(int or float) ，就必须使用int()或float()来将str转换为int或float

## 2、写入文件

### 保存数据最简单的方式之一就是将数据保存到文件中。通过将输出写入文件，即便关闭包含程序输出的终端窗口，这些输出也依然存在，我们可以在程序运行结束后再去查看程序的输出，也可与别人分享，还可编写程序将文件再次读入内存中进行其他的操作

### 2.1 写入空文件

#### 使用open()函数将文本写入文件时，需要向open()函数提供一个表示要写入文本的文件，告诉Python你要写入打开的文件，如果文件不存在，Python会自动创建该文件。当提供的文件不带路径时，会在当前工作工作路径下创建相应的文件；当提供的文件带路径时，会在相应到的路径下产生对应的文件

```python
with open('programming.txt','w') as file:
    file.write('I love Python Programming Language!')
```

#### 【注】：
##### ①在上述代码中，‘w’表示写入模式。当要写入文本的文件已存在(当前工作目录或是指定的工作目录)时，Python会先删除原先存在的文件，再生成新的同名文件，并将文本写入到新产生的文件中。查看open()函数了解更多的模式：
##### ②Python只能将字符串写入到文本文件，要把数值数据保存到文本文件，需要将数值数据先转换( str() )为字符串，再进行存储

```python
help(open)
```

### 2.2 写入多行

#### write() 不会在我们写入的文本末尾自动添加换行符，所以需要手动添加换行符（'\n'），还可以使用制表符、空格等来设置这些格式

```python
with open('PythonLanguage.txt','w') as writer_file:
    writer_file.write('Python is a very good Programming!\n')
    writer_file.write('I love it very much!')
```

### 2.3 附加到文件

#### 有时候，我们并不是想重新创建一个新的文件，只是想向某个指定的文件中加入一些额外的信息，而不是覆盖原有的内容。此时，我们可采用附加模型（'a'）打开文件，新写入的内容会添加到已有内容的后面；若指定的文件不存在，Python会自动创建一个空文件，并把文本写入到该文件中。ex：向已存在的Programming.txt文件中加入一些新的内容：

```python
with open('Programming.txt','a') as file_obj:
    file_obj.write('I love programming!\n')
    file_obj.write('I love using Python to train deep learning algorithms!')
```

#### 【注】：将已存在的字符串列表保存到文件中

# 二、数据保存与恢复——序列化与反序列化

## Python中，常用来对数据进行序列化和反序列化操作的模块有pickle和json，他们均有dump, dumps, load, loads方法，不同之处在于：
### json：用于字符串（来自不同平台和不同语言）和Python数据类型进行转换
### pickle：用于Python特有的类型与Python的数据类型之间的转换（所有Python的数据类型）

```python
import json,pickle
```

### ①dump, dumps

#### dump和dumps都是对数据进行序列化，不同的是：dumps只进行序列化，并不对序列化的结果进行保存，只需要一个参数——需要序列化的数据；dump进行序列化并将序列化的结果保存到指定的文件中，需要两个参数——需要序列化的数据和序列化数据的保存地址。ex：

```python
test_dict = {
    'name':'QingCai',
    'start_time':'2014-12-8',
    'address':'Kunming',
    'CEO':'段国涛'
}
```

```python
pickle.dumps(test_dict)              # 序列化后为字节类型
json.dumps(test_dict)                # 序列化为一个字符串
```

```python
with open('pickle_test.dat','wb') as file:          # 需要使用'wb'模式
    pickle.dump(test_dict,file)
```

```python
with open('json_test.json','w') as file:          # 需要使用'w'模式
    json.dump(test_dict,file)
```

### ②load, loads

#### load和loads都是对数据进行反序列化操作，不同的是：loads只对数据进行反序列化，并不进行数据的读入操作load文件中的数据读入并进行反序列化后返回。ex：

```python
pickle.loads(pickle.dumps(test_dict))
json.loads(json.dumps(test_dict))
```

```python
with open('pickle_test.dat','rb') as file:
    pickle_test = pickle.load(file)
    print(pickle_test)
```

```python
with open('json_test.json') as file:
    json_test = json.load(file)
    print(json_test)
```
